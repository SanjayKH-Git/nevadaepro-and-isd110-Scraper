```python
nevada_search_url = "https://nevadaepro.com/bso/view/search/external/advancedSearchBid.xhtml?openBids=true"
nevada_search_resp = requests.get(nevada_search_url,  headers={'User-Agent': 'Mozilla/5.0'})
nevada_search_soup = Bs(nevada_search_resp.text, "html.parser")
dom = etree.HTML(str(nevada_search_soup))
nevada_search_soup.prettify()
```

```python
bid_collection = {}
rows = nevada_search_soup.find('tbody', id='bidSearchResultsForm:bidResultId_data').find_all('tr')
for row in rows:
    # Extracting First page
    bid_solicitation_col = row.select_one("a[href*='/bidDetail.sdo?']")
    bid_solicitation_name, bid_solicitation_url = bid_solicitation_col.text, "https://nevadaepro.com" + bid_solicitation_col.get('href')
    td_row = row.find_all('td')
    buyer = td_row[5].text  
    description = td_row[6].text
    bid_opening_date = td_row[7].text
    
    # Extracting header information from bid info detail page
    header_information = {}
    
    
    bid_collection[bid_solicitation_name] = {
        'Bid Solicitation #': bid_solicitation_name,
        'Buyer': buyer,
        'Description': description,
        'Bid Opening Date': bid_opening_date,   
        'header information': {},     
        }

bid_collection
    # print(bid_solicitation_name, bid_solicitation_url, buyer, description, bid_opening_date)
    
```
